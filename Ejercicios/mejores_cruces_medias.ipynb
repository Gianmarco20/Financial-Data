```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
def mejores_cruces(ticker):
    data = yf.download(tickers=ticker, auto_adjust=True)
    data['ma_20'] = data['Close'].rolling(19).mean()
    data['ma_50'] = data['Close'].rolling(50).mean()
    data['ratio'] = data['ma_20'] / data['ma_50']
    data['decision'] = np.where((data['ratio'].shift(2)<1) & (data['ratio'].shift()>=1), 'Buy', np.where((data['ratio'].shift(2)>=1) & (data['ratio'].shift()<1), 'Sell', 'Stay'))
    data['state'] = np.where(data['ratio'].shift()>=1, 'In', np.where((data['ratio'].shift(2)>=1) & (data['ratio'].shift()<1), 'Sell', 'Stay'))
    return data
mejores_cruces('AAPL')
```
