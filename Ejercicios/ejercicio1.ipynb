```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import datetime as dt
```

```python
data_alycbur = pd.read_excel('/Applications/Python 3.10/Data Science Coursera/Operaciones Alycbur.xlsx', header=3)
clientes = pd.read_excel('/Applications/Python 3.10/Data Science Coursera/Listado Clientes Operados.xlsx')
operaciones = pd.read_excel('/Applications/Python 3.10/Data Science Coursera/Listado Categorias Operaciones.xlsx')

data_alycbur['Empresa'] = 'ALYCBUR'
data_alycbur.head()
```

```python
operaciones_alycbur = data_alycbur['Tipo de operación'].unique().tolist()
filtro_operaciones = [operacion for operacion in operaciones_alycbur if 'Subasta' in operacion]
filtro_operaciones
```

```python
data_alycbur = data_alycbur[data_alycbur['Tipo de operación'].isin(filtro_operaciones)]
data_alycbur = data_alycbur.drop(columns=['Gastos'])
data_alycbur = data_alycbur[(data_alycbur['Denominación']!='ALYCBUR S.A.') & (data_alycbur['Denominación']!='FINARES SA') & (data_alycbur['Denominación']!='ALYCBUR SA') & (data_alycbur['Denominación']!='ADCAP BALANCEADO VIII') & (data_alycbur['Denominación']!='ALYCBUR FONDO COMÚN DE INVERSIÓN ABIERTO PYMES') & (data_alycbur['Denominación']!='GOLDMAN, EDUARDO ELIAS')]
data_alycbur['CUIT'] = data_alycbur['CUIT'].str.replace('-', '').astype(int)
data_alycbur
```

```python
clientes
```

```python
data_alycbur_actualizada = pd.merge(data_alycbur, clientes[['CUIT', 'Oficial']], on='CUIT', how='left')
data_alycbur_actualizada
```

```python
columnas = {
    'Boleto': 'Boleto / Operación',
    'Cuenta': 'N° Cliente',
    'Denominación': 'Cliente',
    'Precio': 'Tasa',
    'Aranceles': 'Gastos',
    'Bruto': 'Bruto_alycbur'
}

data_alycbur_actualizada = data_alycbur_actualizada.rename(columns=columnas)
data_alycbur_actualizada
```

```python
data_alycbur_actualizada_2 = pd.merge(data_alycbur_actualizada, clientes[['CUIT', 'Ubicación']], on='CUIT', how='left')
data_alycbur_actualizada_2
```

```python
data_alycbur_actualizada_3 = pd.merge(data_alycbur_actualizada_2, clientes[['CUIT', 'Grupo / Cliente']], on='CUIT', how='left')
data_alycbur_actualizada_3
```

```python
operaciones
```

```python
data_alycbur_final = pd.merge(data_alycbur_actualizada_3, operaciones[['Tipo de operación', 'Categorias']], on='Tipo de operación', how='left')
data_alycbur_final = data_alycbur_final.drop(columns=['Tipo de operación'])
data_alycbur_final
```

```python
data_alycbur_final = data_alycbur_final.rename(columns={'Categorias': 'Tipo de operación'})
data_alycbur_final
```

```python
data_alycbur_final[~data_alycbur_final['Tipo de cambio'].isnull()]
```

```python
data_alycbur_final['Tipo de cambio'] = data_alycbur_final['Tipo de cambio'].astype(str)
```

```python
data_alycbur_final['Tipo de cambio'] = data_alycbur_final['Tipo de cambio'].str.extract(r'ARS\s(\d+\,?\d*)', expand=False)
data_alycbur_final['Tipo de cambio'] = data_alycbur_final['Tipo de cambio'].str.replace(',', '.').astype(float)
data_alycbur_final['Tipo de cambio'] = data_alycbur_final['Tipo de cambio'].fillna(0)
```

```python
data_alycbur_final = data_alycbur_final.rename(columns={'Tipo de cambio': 'TC usado', 'Cantidad': 'Bruto (no usar)'})
data_alycbur_final
```

```python
data_alycbur_final['Tasa'] = data_alycbur_final['Tasa'] / 100
data_alycbur_final['Bruto'] = data_alycbur_final.apply(lambda row: row['Bruto (no usar)'] * row['TC usado'] if row['Tipo de operación'] == 'PAGARÉ' else row['Bruto (no usar)'], axis=1)
data_alycbur_final['Tasa']
```

```python
data_alycbur_final['Neto'] = data_alycbur_final['Neto'].str.extract(r'ARS -(\d+\,?\d*)', expand=False)
data_alycbur_final['Neto'] = data_alycbur_final['Neto'].str.replace(',', '.').astype(float)
data_alycbur_final['Neto']
```

```python
data_alycbur_final['Gastos'] = data_alycbur_final['Gastos'].str.extract(r'ARS (\d+\,?\d*)', expand=False)
data_alycbur_final['Gastos'] = data_alycbur_final['Gastos'].str.replace(',', '.').astype(float)
data_alycbur_final['Gastos']
```

```python
data_alycbur_final['Interes'] = data_alycbur_final['Bruto (no usar)'] - data_alycbur_final['Bruto_alycbur']
data_alycbur_final['Interes']
```

```python
data_alycbur_final['Facturación'] = data_alycbur_final['Interes'] + data_alycbur_final['Gastos']
data_alycbur_final['Facturación']
```

```python
data_alycbur_final['Vto. instrumento'] = pd.to_datetime(data_alycbur_final['Vto. instrumento'], format='%Y/%m/%d')
data_alycbur_final['Días Promedio'] = (data_alycbur_final['Vto. instrumento'] - data_alycbur_final['Concertación'])
data_alycbur_final['Días Promedio']
```

```python
data_alycbur_final['%G'] = data_alycbur_final['Gastos'] / data_alycbur_final['Bruto (no usar)']
data_alycbur_final['%G']
```
